```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Parameters
mean = 100
std_dev = 4
data_size = 100000

```

```python
# Generate normal distribution data
data = np.random.normal(mean, std_dev, data_size)

```

```python
# Create histogram data
counts, bin_edges = np.histogram(data, bins=100, density=True)

```

```python
# Calculate the center of each bin for bar plot
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

```

```python
# Set up the bar plot
plt.figure(figsize=(10, 6))
plt.bar(bin_centers, counts, width=bin_edges[1] - bin_edges[0], color='pink', alpha=0.6, label='Histogram')
```

```python
# This graph shows the bell curved shape
# The spread of the data is controlled by the standard deviation.
# About 68% of the data falls within one standard deviation,and 95% within two standard deviations.
#The mean of 100 serves as the center of the distribution, highlighting that the majority of values are near this point.
```
